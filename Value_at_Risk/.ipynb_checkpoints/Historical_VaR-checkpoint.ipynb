# Historical Simulation

```python
import pandas as pd
import numpy as np
import scipy.stats as sp
import matplotlib.pyplot as plt
```

### Data

```python
df = pd.read_excel(r'Data.xlsx', sheetnames = 'Ex 1.1')
df = df.set_index('Date') 
df['Log Rtn'] = np.log(df['SP 500']/df['SP 500'].shift())
df.head()
```

```python
fig, ax1 = plt.subplots(figsize = (15,8))

ax2 = ax1.twinx()
ax1.plot(df.index, df['SP 500'], 'black')
ax2.plot(df.index, df['Log Rtn'], 'lightgrey')

ax1.set_xlabel('X data')
ax1.set_ylabel('S&P', color='black')
ax2.set_ylabel('Log Ret', color='black')

plt.show()
```

### 1.1 Equally Weighted Historical and Linear VaR

```python
df_1_1 = df.copy()

significance_level = 0.01
nobs1 = 500
nobs2 = 2000

df_1_1 = df_1_1.dropna(axis=0)

#historical VaR
df_1_1['Hist_n500'] = -df_1_1['Log Rtn'].rolling(nobs1).quantile(significance_level, interpolation='linear')
df_1_1['Hist_n2000'] = -df_1_1['Log Rtn'].rolling(nobs2).quantile(significance_level, interpolation='linear')

#Normal VaR
df_1_1['Norm_n500'] = sp.norm.ppf(1 - significance_level, loc=0, scale=1)*df_1_1['Log Rtn'].rolling(nobs1).std() - df_1_1['Log Rtn'].rolling(nobs1).mean()
df_1_1['Norm_n2000'] = sp.norm.ppf(1 - significance_level, loc=0, scale=1)*df_1_1['Log Rtn'].rolling(nobs2).std() - df_1_1['Log Rtn'].rolling(nobs2).mean()

#difference
df_1_1['Diff_n500'] = df_1_1["Hist_n500"] - df_1_1["Norm_n500"]
df_1_1['Diff_n2000'] = df_1_1["Hist_n2000"] - df_1_1["Norm_n2000"]
```

```python
plt.plot(df_1_1[['Hist_n500', 'Hist_n2000']])
```

### Exponential Weighting of Return Distributions

With equal weight, the ordering of observations is irrelevant.
To adapt the hisotical VaR so that it no longer weights data equally, we assign an exponential weight to the probability of each return in its distribution. The return weights, from the most recent, will be:
 $$ \lambda(1 - \lambda), \lambda^{2}(1 - \lambda), \lambda^{3}(1 - \lambda), \lambda^{4}(1 - \lambda), ..  $$
